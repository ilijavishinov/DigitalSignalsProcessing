```python
import pandas as pd
import matplotlib.pyplot as plt
from random import randint
import os
from matplotlib import gridspec
plt.style.use(['dark_background'])
plt.rcParams["figure.figsize"] = (10,7)
```

# cudb

```python
for i in range(5):

    random_patient_id = randint(1,35)
    patient_id_name = f'0{random_patient_id}' if random_patient_id<10 else f'{random_patient_id}'

    sampled_dataset = pd.read_csv(f'data\physionet_dbs\cudb\samples\cu{patient_id_name}.csv')
    sampled_dataset.columns = ['sample_idx', 'ecg']

    random_samples_position = randint(1,100000)
    plt.plot(
            sampled_dataset.iloc[random_samples_position:random_samples_position+1000]['sample_idx'],
            sampled_dataset.iloc[random_samples_position:random_samples_position+1000]['ecg'],
    )
    plt.title(f'cu{patient_id_name}')
    plt.show()
```

# mitdb

```python
for i in range(5):

    mitdb_sampled_patients = list(os.listdir('D:\FINKI\8_dps\Project\DATA\physionet_dbs\mitdb\samples'))
    random_file_position = randint(0,len(mitdb_sampled_patients)-1)
    file_name = mitdb_sampled_patients[random_file_position]

    sampled_dataset = pd.read_csv(f'D:\FINKI\8_dps\Project\DATA\physionet_dbs\mitdb\samples\{file_name}')
    sampled_dataset.columns = ['sample_idx', 'MLII', 'V1']

    gs = gridspec.GridSpec(2, 1)
    ax1 = plt.subplot(gs[0, 0])
    ax2 = plt.subplot(gs[1, 0])

    plt.title(f'{file_name.replace(".csv","")} MLII signal (top) vs V1 signal (bottom)')


    random_samples_position = randint(1,100000)
    ax1.plot(
            sampled_dataset.iloc[random_samples_position:random_samples_position+1000]['sample_idx'],
            sampled_dataset.iloc[random_samples_position:random_samples_position+1000]['MLII'],
    )

    ax2.plot(
            sampled_dataset.iloc[random_samples_position:random_samples_position+1000]['sample_idx'],
            sampled_dataset.iloc[random_samples_position:random_samples_position+1000]['V1'],
    )


    plt.show()
```
